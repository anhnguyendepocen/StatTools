# Polynomial Regression

We will show how to use the `PolynomialRegression` class in the MLTools `glm` module to fit a polynomial curve through a 2D scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", palette="muted", color_codes=True)

from mltools.glm import PolynomialRegression
```

```python
# Set NumPy random number generator seed for replicability
np.random.seed(100)
```

## Create Some Artificial Data

```python
n = 100

# Actual model coefficients
a = 5
b = -3
c = 1

x = np.random.uniform(0, 10, n)
y = a + b * x + c * (x ** 2) + np.random.normal(0, 20, n)

# Scatter plot of the raw data
plt.figure()
plt.scatter(x, y, c="b", alpha=0.7)

# Plot of the true curve
xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()
t = np.linspace(xmin, xmax, num=100)
plt.plot(t, a + b * t + c * (t ** 2), c="k")
plt.xlim((xmin, xmax))
plt.ylim((ymin, ymax))

plt.title(f"y = ${a} {b} x + {c} x^2$")
plt.show()
plt.close()
```

## Linear Model

```python
%time
model = PolynomialRegression(deg=1).fit(x, y)
```

```python
# Plot the regression line
plt.figure()
plt.scatter(x, y, c="b", alpha=0.7)
model.fit_plot(lw=3, c="r")
plt.title(f"${model.poly_str()}$")
plt.show()
plt.close()
```
