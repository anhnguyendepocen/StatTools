# Polynomial Regression

We will show how to use the `PolynomialRegression` class in the MLTools `glm` module to fit a polynomial curve through a 2D scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", palette="muted", color_codes=True)

from mltools.glm import PolynomialRegression
```

```python
# Set NumPy random number generator seed for replicability
np.random.seed(100)
```

## Create Some Artificial Data

```python
n = 100
a = 5
b = -3
c = 1

x = np.random.uniform(0, 10, n)
y = a + b * x + c * (x ** 2) + np.random.normal(0, 10, n)
```

```python
%time
model = PolynomialRegression(deg=2).fit(x, y)
```

```python
# Get coefficients of the model
a_hat, b_hat, c_hat = model.coef

# Plot the regression curve
plt.figure()
plt.scatter(x, y, c="b", alpha=0.7)
model.fit_plot(lw=3, c="r", ls="--")
plt.title(f"Prediction: $y = {a_hat:.3f} {'+' if b_hat >= 0 else ''} {b_hat:.3f}x {'+' if c_hat >= 0 else ''} {c_hat:.3f}x^2$")
plt.show()
plt.close()
```
