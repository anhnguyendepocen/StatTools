# Simulating Itô Diffusions

An Itô diffusion is a continuous-time stochastic process $X$ which satisfies a stochastic differential equation (SDE) of the form
$$
dX(t) = b(X(t)) dt + \sigma(X(t)) dB(t),
$$
where $B$ is standard Brownian motion.
The function $b$ is called the *drift coefficient*, and the function σ is called the *diffusion coefficient*.

StatTools provides the `ItoDiffusion` class in its `simulation` module for simulating sample paths of Itô diffusions using the Euler-Maruyama method.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", palette="muted", color_codes=True)

from stattools.simulation import BrownianMotion, GeometricBM, OrnsteinUhlenbeck
```

## Standard Brownian Motion

```python
plt.figure()
for seed in range(3):
    bm = BrownianMotion(random_state=seed)
    bm.plot(end=1.0)
plt.show()
plt.close()
```

## Brownian Motion with Drift

```python
plt.figure()
for seed in range(3):
    bm = BrownianMotion(mu=2, sigma=0.7, x0=2, random_state=seed)
    bm.plot(end=1.0)
plt.show()
plt.close()
```

## Geometric Brownian Motion

```python
plt.figure()
for seed in range(3):
    gbm = GeometricBM(mu=0.15, sigma=0.20, step=1e-2, random_state=seed)
    gbm.plot(end=5.0)
plt.show()
plt.close()
```

## Ornstein-Uhlenbeck Process

```python
plt.figure()
for seed in range(3):
    ou = OrnsteinUhlenbeck(theta=0.8, mu=1.5, sigma=0.09, step=1e-2, random_state=seed)
    ou.plot(end=8.0)
plt.show()
plt.close()
```
