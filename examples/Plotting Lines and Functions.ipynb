# Plotting Lines and Functions

We will show how to use the `abline` and `func_plot` functions in the StatTools `visualization` module to plot lines and functions.

```python
import numpy as np
import scipy.stats as st

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", palette="muted", color_codes=True)

from stattools.visualization import abline, func_plot
```

## Plotting Lines

```python
plt.figure()
abline(2, 3, x_min=-1, x_max=2, label="$y=2+3x$")
abline(0, 1, label="$y=x$")
abline(6, -2, label="$y=6-2x$")
plt.legend(loc="upper left", frameon=True, shadow=True)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
plt.close()
```

## Plotting Functions

```python
plt.figure()
func_plot(np.log1p, x_min=0, x_max=2, label="$y=\log(1+x)$")
func_plot(np.sin, label="$y=\sin(x)$")
func_plot(np.cos, label="$y=\cos(x)$")
plt.legend(loc="best", frameon=True, shadow=True)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
plt.close()
```

```python
dist = st.norm(loc=0, scale=1)

x = dist.rvs(size=1000, random_state=0)

plt.figure()
plt.hist(x, bins="auto", density=True, alpha=0.7)
func_plot(dist.pdf, c="r", label="$N(0, 1)$ Density")
plt.legend(loc="best", frameon=True, shadow=True)
plt.xlabel("$x$")
plt.ylabel("Density")
plt.show()
plt.close()
```
